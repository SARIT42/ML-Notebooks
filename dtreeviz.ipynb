```python
import graphviz.backend as be
```

```python
from sklearn.datasets import *
from dtreeviz.trees import *
from IPython.display import Image,display_svg,SVG
```

```python
clas=tree.DecisionTreeClassifier(max_depth=3)
iris=load_iris()
```

```python
x_train=iris.data
y_train=iris.target
clas.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clas)
```

DTREEVIZ

```python
viz=dtreeviz(clas,
             x_train,
             y_train,
             target_name='flower type',
             feature_names=iris.feature_names,
             class_names=['setosa','versicolor','virginica'],
             scale=2)
viz
```

FOR REGRESSION TREE

```python
boston=load_boston()
```

```python
reg=tree.DecisionTreeRegressor(max_depth=2)
import pandas as pd
import numpy as np


data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
x_train = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
y_train = raw_df.values[1::2, 2]
reg.fit(x_train,y_train)

viz=dtreeviz(reg,
             x_train,
             y_train,
             target_name='price',
             feature_names=boston.feature_names,
             scale=2
             #orientation='LR'
             )
```

```python
viz
```

SHOW PREDICTION PATH

```python
x_train1=iris.data
y_train1=iris.target
clas.fit(x_train1,y_train1)

x=iris.data[np.random.randint(0,len(iris.data)),:]

viz=dtreeviz(clas,
             x_train1,
             y_train1,
             feature_names=iris.feature_names,
             class_names=['setosa','versicolor','virginica'],
             orientation='LR',
             show_node_labels=True,
             X=x)
viz
```
