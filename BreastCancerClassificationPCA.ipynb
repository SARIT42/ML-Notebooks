```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
df=pd.DataFrame(cancer['data'],columns=(cancer['feature_names']))
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
sdf=scaler.fit_transform(df)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(sdf,cancer['target'],test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

PCA

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
sdf=pca.fit_transform(sdf)
```

```python
sdf.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(sdf,cancer['target'],test_size=0.25,random_state=42)
```

```python
lor=LogisticRegression()
lor.fit(x_train,y_train)
```

```python
lor.score(x_test,y_test)
```
